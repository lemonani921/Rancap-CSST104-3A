**1. Creating dataset**

```python
import pandas as pd

data = {
    'student_id': range (1,21),
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female' ],
    'test_date': pd.date_range(start = '2024-02-21', periods = 20),
    'study_hours': [2,1,5,3,4,2,1,2,3,4,
                    5,1,2,3,2,1,2,2,1,1],
    'test_score': [81,76,92,83,89,82,78,82,84,88,
                   82,78,95,88,91,81,77,83,85,90],
    'attendance': ['Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent', 'Present', 'Absent',]

}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index = False)
```

**2. Data Loading and Initial Examination**

```python
df = pd.read_csv('student_performance.csv')

print(df.head())

print(df.isnull().sum())
```

**3. Exploratory Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score', 'study_hours']].describe()

plt.figure(figsize = (12, 5))
plt.subplot(1, 2, 1)
sne.histplot(df['test_score'], bins = 15, kde = True)
plt.title('Distribution of Test Score')

plt.subplot(1, 2, 2)
sne.histplot(df['study_hours'], bins = 15, kde = True)
plt.title('Distribution of Study Hours')

plt.figure(figsize = (8 , 6))
sne.scatterplot(data = df, x = 'study_hours', y = 'test_score')
plt.title('Scatter Plot of Study Hours vs. Test Score')
```

**4. Data Visualization**

```python
plt.figure(figsize = (6, 4))
sne.countplot(data = df, x = 'gender')
plt.title('Gender Distribution')

plt.figure(figsize = (10, 6))
sne.lineplot(data = df, x = 'test_date', y = 'test_score')
plt.title('Trends in Test Scores Over Time')

correlation_matrix = df[['study_hours', 'test_score', 'attendance']].corr()

plt.figure(figsize = (8, 6))
sne.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('Heat Map')
```
