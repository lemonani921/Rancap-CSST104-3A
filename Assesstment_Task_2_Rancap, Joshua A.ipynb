**1. Data Loading**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sne


#Loading the dataset to the pandas DataFrame
df = pd.read_csv('hardwareStore.csv')
```

**2. Initial Data Analysis**

```python
#Display basic statistics of the dataset, including mean, median, and standard deviation of numeric columns.
statistic = df.describe()
print(statistic)

#Identify the unique categories and their counts.
print(df.head())
print(df.isnull().sum())
```

**3. Category Analysis**

```python
plt.figure(figsize = (12, 5))
sne.histplot(df['CATEGORY_NAME'], bins = 15, kde = True)
plt.title('Distribution of Category Name')

product_revenue = df.groupby('CATEGORY_NAME')['QUANTITY'].sum()
highest_revenue_product = product_revenue.idxmax()

print("The category that has the highest number of products: " + highest_revenue_product)
```

**4. Cost and Price Analysis**

```python
average_Scost = df['STANDARD_COST'].mean()
average_Lprice = df['LIST_PRICE'].mean()

print("Average standard cost: ", average_Scost)
print("Average list price: ", average_Lprice)

standard_cost = df.groupby('PRODUCT_NAME')['STANDARD_COST'].sum()
list_price = df.groupby('PRODUCT_NAME')['LIST_PRICE'].sum()

highest_standard_cost =  standard_cost.idxmax()
highest_list_price =  list_price.idxmax()

print("The product that have highest standard cost: ", highest_standard_cost)
print("The product that have highest list price: ", highest_list_price)
```

**5. Location Analysis**

```python
product = df.groupby('CITY')['QUANTITY'].sum()
city_highprod = product.idxmax()

print("The city that have highest quantity of products: ", city_highprod)
```

**6. Insights and Reporting**

```python
report = """
                                      *** Data Analysis Report ***

                                      *** Summary Statistics ***

                                The category that has the highest number of products: {}
                                Average standard cost: {:.2f}
                                Average list price: {:.2f}
                                The product that have highest standard cost: {}
                                The product that have highest list prrice: {}
                                The city that have highest quantity of products: {}

{}





""".format(highest_revenue_product, average_Scost, average_Lprice, highest_standard_cost, highest_list_price, city_highprod,statistic)


plt.figure(figsize = (10, 6))
sne.barplot(data = df, x = 'CATEGORY_NAME', y = 'STANDARD_COST', color = 'gray')
plt.title('Standard Cost of Categories of Products')

plt.figure(figsize = (10, 6))
sne.barplot(data = df, x = 'CATEGORY_NAME', y = 'QUANTITY', color = 'purple')
plt.title('Quantities of the Categories of Products')

plt.figure(figsize = (10, 6))
sne.lineplot(data = df, x = 'COUNTRY_ID', y = 'QUANTITY')
plt.title('Countries with Highest Quantities of Products')


print(report)
```
