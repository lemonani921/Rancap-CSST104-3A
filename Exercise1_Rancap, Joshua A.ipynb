**1. Importing Libraries**

```python
import pandas as pd # Data Manipulation
import matplotlib.pyplot as plt # Data Visualization
```

**2. Loading Datasets**

```python
data = pd.read_csv('sales_data_2.csv')
```

**3. Data Loading and Initial Analysis**

```python
statistic = data.describe()
```

**4. Data Manipulation**

```python
product_revenue = data.groupby('Product')['Revenue'].sum()
highest_revenue_product = product_revenue.idxmax()
average_quantity_per_day = data['Quantity Sold'].mean()

print(product_revenue)
print(highest_revenue_product)
print(average_quantity_per_day)
```

**5. Data Visualization**

```python
product_revenue.plot(kind = 'bar', color = 'purple')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

**6. Analysis Report**

```python
report = """
Analysis Report

1. Data Loading and Initial Analysis:
   Basic statistics of the dataset:
   {}
2. Data Manipulation:
   Total revenue for each product:
   {}
   Product with the highest revenue: {}
   Average quantity sold per day: {:.2f}
3. Data Visualization
   Bar chart displaying revenue for each product is attached.
4. Insights
   -The product {} generates the highest revenue
   -On average, {:.2f} units sold per day.

""".format(statistic, product_revenue, highest_revenue_product, average_quantity_per_day, highest_revenue_product, average_quantity_per_day)

print(report)
```

**7. Saving the Report to a File**

```python
with open('report.txt', 'w') as f:
  f.write(report)
```
